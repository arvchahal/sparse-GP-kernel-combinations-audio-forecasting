```python
import jax.numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.load("output_file.npy", allow_pickle=True)
print(data)
```

```python
# Split the data into groups
group_1 = data[:5]
group_2 = data[5:10]
group_3 = data[10:]

def extract_names_and_values(group):
    names = [row[0].split('_en')[0] for row in group]  # Extract names before '_en'
    values = np.array([row[1:].astype(float) for row in group])  # Convert remaining values to floats
    return names, values

dates = pd.date_range(start="2015-07-01", periods=803, freq="D")
```

```python
# Extract data for Group 1
names_1, values_1 = extract_names_and_values(group_1)

# Plot Group 1
plt.figure(figsize=(10, 6))
for i, name in enumerate(names_1):
    plt.plot(dates, values_1[i], label=name)
plt.title("Major Tech Companies")
plt.xlabel("Date")
plt.ylabel("Views")
plt.legend(loc="upper left")
plt.show()
```

```python
# Extract data for Group 2
names_2, values_2 = extract_names_and_values(group_2)

# Plot Group 2
plt.figure(figsize=(10, 6))
for i, name in enumerate(names_2):
    plt.plot(dates, values_2[i], label=name)
plt.title("Group 2")
plt.xlabel("Date")
plt.ylabel("Views")
plt.legend(loc="upper left")
plt.show()
```

```python
# Extract data for Group 3
names_3, values_3 = extract_names_and_values(group_3)

# Plot Group 3
plt.figure(figsize=(10, 6))
for i, name in enumerate(names_3):
    plt.plot(dates, values_3[i], label=name)
plt.title("Group 3")
plt.xlabel("Date")
plt.ylabel("Views")
plt.legend(loc="upper left")
plt.show()
```
